# Funciones de distribución de probabilidades 
### Distribución Uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
    
```

```python
a=1
b=100
c=1000000
data = np.random.uniform(a,b,c)
```

```python
%matplotlib inline
plt.hist(data)
```
