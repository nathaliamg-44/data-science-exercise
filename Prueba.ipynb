```python
def creayusardummy(df, columna):
    dummy = pd.get_dummies(df[columna],prefix= columna)
    df = df.drop([columna], axis =1)
    df = pd.concat([df,dummy], axis=1)
    #df.to_excel(mainpath+"/"+"dummiesprueba.xlsx")... opcion de guardar con la funcion
    return df
```

```python
creayusardummy(pruebadumy, "genero")
```
